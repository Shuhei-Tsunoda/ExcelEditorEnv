# 列が階層的なExcelファイルを読み込む

```python
import numpy as np
import pandas as pd
```

```python
major_index_raw = 2
minor_index_raw = 3
start_row = min(major_index_raw, minor_index_raw)
start_col = 1
end_col = 21
df = pd.read_excel('resource/2017_active_safety_edited.xlsx',
                   header=None,
                   skiprows=[*range(start_row)],
                   usecols=[*range(start_col, end_col+1)])
df.head(3)
```

```python
major_index = 0
df.iloc[major_index, 0]= 'id'
for i in range(1, len(df.columns)):
    if pd.isna(df.iloc[major_index, i]):
        df.iloc[major_index, i] = df.iloc[major_index, i-1]
minor_index = 1
for i in range(len(df.columns)):
    if not pd.isna(df.iloc[minor_index, i]):
        df.iloc[major_index, i] = f'{df.iloc[major_index, i]}|{df.iloc[minor_index, i]}'.replace('\n', '')
df = df.drop(1)

df.head(3)
```

```python
df.columns = df.iloc[major_index]
df = df.drop(major_index)
del major_index, minor_index
df.head(3)
```

```python
df = df.set_index('id')
df.head(3)
```

```python
df.index = df.index.astype(np.int32)
df.head(3)
```
